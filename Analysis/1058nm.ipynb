```python
def biexpDecay(x, a, b, c, d, e):
    return (a*np.exp(-x/b) + c*np.exp(-x/d) +e)

def expDecay(x, a, b, c):
    return -a*np.exp(-x/b) + c

def normExpDecay(x, a, b, c):
    return a*np.exp(-x/b) + c

def stretchedExpDecay(x, a, b, c, d):
    return a*np.exp(-((x/b)**d)) + c
```

```python
def fitbiInvRec(data):
    popt, pcov = curve_fit(biexpDecay,data.index.values, data['Real'], bounds = ([-10, 0, -10, 0, -20], [0, 20, 0, 20, 20]))
    errs = np.sqrt(np.diag(pcov))
    if popt[1] <= popt[3]:
        T1a = popt[1]
        T1aerr = errs[1]
        T1berr = errs[3]
        T1b = popt[3]
    else:
        T1a = popt[3]
        T1aerr = errs[3]
        T1berr = errs[1]
        T1b = popt[1]
    errs = np.sqrt(np.diag(pcov))
    return T1a, T1aerr, T1b, T1berr

def fitInvRec(data):
    popt, pcov = curve_fit(expDecay, data.index.values, data['Real'])
    T1 = popt[1]
    errs = np.sqrt(np.diag(pcov))
    T1err = errs[1]
    return T1, T1err

def fitStretchedExp(data):
    popt, pcov = curve_fit(stretchedExpDecay, data.index.values, data['Real'], bounds = ([0, 0, -5, 1], [10, 10, 5, 3]))
    T2 = popt[1]
    errs = np.sqrt(np.diag(pcov))
    T2err = errs[1]
    Stretch = popt[3]
    StErr = errs[3]
    return T2, T2err, Stretch, StErr
```

```python
def getT1Data(T1Dict):
    T1sa = []
    T1sb = []
    T1Powers = []
    T1aErrors = []
    T1bErrors = []
    
    for data in T1Dict.keys():
        if T1Dict[data].power == 0:
            T18k, T18kerr = fitInvRec(T1Dict[data])
            T1sa.append(T18k)
            T1sb.append(T18k)
            T1aErrors.append(T18kerr)
            T1bErrors.append(T18kerr)
            

        T1a, T1aerr, T1b, T1berr = fitbiInvRec(T1Dict[data])
        if T1a < T1b:
            T1sa.append(T1a)
            T1sb.append(T1b)
            T1aErrors.append(T1aerr)
            T1bErrors.append(T1berr)
        else:
            T1sa.append(T1b)
            T1sb.append(T1a)
            T1bErrors.append(T1aerr)
            T1aErrors.append(T1berr)
        T1Powers.append(T1Dict[data].power)
    T1DataFrame = pd.DataFrame()
    T1DataFrame['T1a'] = T1sa
    T1DataFrame['T1a Error'] = T1aErrors
    T1DataFrame['T1b'] = T1sb
    T1DataFrame['T1b Error'] = T1bErrors
    T1DataFrame.index = T1Powers
    T1DataFrame.sort_index()
    return T1DataFrame
```

```python
def getT2Data(T2Dic):
    T2s = []
    T2errs = []
    T2Powers = []
    Stretchs = []
    StErrs = []
    for data in T2Dic:
        T2, T2err, Stretch, StErr = fitStretchedExp(T2Dic[data])
        T2s.append(T2)
        T2Powers.append(T2Dic[data].power)
        Stretchs.append(Stretch)
        StErrs.append(StErr)
        T2errs.append(T2err)
    T2DatFrame = pd.DataFrame()
    T2DatFrame['T2'] = T2s
    T2DatFrame['T2 Error'] = T2errs
    T2DatFrame['Stretch'] = Stretchs
    T2DatFrame['Stretch Error'] = StErrs
    T2DatFrame.index = T2Powers
    T2DatFrame.sort_index()
    return T2DatFrame
        
```
